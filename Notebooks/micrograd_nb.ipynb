```python
from micrograd import Neuron, Layer, MLP

xs = [
    [3., 3.4, 5.5],
    [2., -1., 90.4],
    [8., 4., 6.],
    [4., 6., -8.]
]
```

```python
x = [3., 3.4, 5.5]
y = [-1., 0., -1., 1]
n = MLP(3,[4,4,4])
```

```python
# step iterations 
for k in range(3000):
    # forward pass 
    ypred = n(x) #[n(x) for x in xs]
    loss = sum((ypredi - yi)**2 for yi, ypredi in zip(y,ypred))
    # backwards 
    for w in n.parameters():
        w.grad = 0.0 
    loss.backwards()
    # update 
    for w in n.parameters():
        w.data += - 0.05 * w.grad 
    
    print(k, loss.data)
    
```
